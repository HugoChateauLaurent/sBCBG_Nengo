```python
import nengo
from nengo.utils.matplotlib import rasterplot
from nengo_extras.plot_spikes import (
    cluster, merge, plot_spikes, preprocess_spikes, sample_by_variance)

import nengo_spa as spa
import numpy as np
spa.modules.basalganglia.BasalGanglia.sBCBG = True
import matplotlib.pyplot as plt

import nengo_ocl
```

## Visual input

```python
# Number of dimensions for the Semantic Pointers
dimensions = 64
n_per_dim = 100

sequence = ['RED', 'BLUE', 'YELLOW', 'WHITE', 'BLACK', 'ORANGE'][:5]


class Input:

    def __init__(self, sequence, duration, dt=0.001):
        self.sequence = sequence
        self.index = 0
        self.duration = duration
        self.dt = dt
        self.count = 0
        
    def step(self, t):
        elt = self.sequence[self.index]
        if self.count > self.duration:
            self.count = 0
            self.index = (self.index + 1) % len(self.sequence)
        self.count += self.dt
        return elt
```

## Define the model and run the experiment

```python


with spa.Network() as model:
    
    # Specify the modules to be used
    vision = spa.Transcode(Input(sequence, .2).step, output_vocab=dimensions)    
    motor = spa.State(dimensions, neurons_per_dimension=n_per_dim)
    motor_gur = spa.State(dimensions, neurons_per_dimension=n_per_dim)
    
    spa.modules.basalganglia.BasalGanglia.sBCBG = True
    with spa.ActionSelection() as action_sel_1:
        for elt in sequence:              
            spa.ifmax(spa.dot(vision, spa.sym(elt)),
                    vision >> motor)
         
    spa.modules.basalganglia.BasalGanglia.sBCBG = False
    with spa.ActionSelection() as gur:
        for elt in sequence:              
            spa.ifmax(spa.dot(vision, spa.sym(elt)),
                    vision >> motor_gur)
    
    bcbg_in = [nengo.Probe(pop) for pop in action_sel_1.bg.input]
    CSN_in = [nengo.Probe(pop.neurons, 'input') for pop in action_sel_1.bg.pops['CSN']]
    CSN = [nengo.Probe(pop.neurons) for pop in action_sel_1.bg.pops['CSN']]
    MSN = [nengo.Probe(pop.neurons) for pop in action_sel_1.bg.pops['MSN']]
    STN = [nengo.Probe(pop.neurons) for pop in action_sel_1.bg.pops['STN']]
    GPe = [nengo.Probe(pop.neurons) for pop in action_sel_1.bg.pops['GPe']]
    GPi = [nengo.Probe(pop.neurons) for pop in action_sel_1.bg.pops['GPi']]
    GPi_decoded = [nengo.Probe(pop) for pop in action_sel_1.bg.output]
    
    gur_in = [nengo.Probe(poqp) for pop in gur.bg.input]
    gur_GPi = [nengo.Probe(pop.neurons) for pop in gur.bg.gpi.ea_ensembles]
    gur_out = [nengo.Probe(pop) for pop in gur.bg.output]
    
    
    
    p_vision = nengo.Probe(vision.output)
    p_motor = nengo.Probe(motor.output)
    p_motor_gur = nengo.Probe(motor_gur.output)
    
    
```

```python
with nengo.Simulator(model) as sim:
    sim.run(2)
```

## Plot results
