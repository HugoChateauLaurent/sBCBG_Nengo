## Visual input

```python
# Number of dimensions for the Semantic Pointers
dimensions = 32
n_per_dim = 50

sequence = ['Flee', 'Flight', 'Fight'][:3]


class Input:

    def __init__(self, sequence, duration, dt=0.001):
        self.sequence = sequence
        self.index = 0
        self.duration = duration
        self.dt = dt
        self.count = 0
        
    def step(self, t):
        elt = self.sequence[self.index]
        if self.count > self.duration:
            self.count = 0
            self.index = (self.index + 1) % len(self.sequence)
        self.count += self.dt
        return elt if np.random.random() < .6 else np.random.choice(self.sequence)
```

## Define the model and run the experiment

```python
with spa.Network() as net:
    
    # Specify the modules to be used
    vision = spa.Transcode(Input(sequence, .2).step, output_vocab=dimensions)    
    motor_normal = spa.State(dimensions, neurons_per_dimension=n_per_dim)
    motor_PD = spa.State(dimensions, neurons_per_dimension=n_per_dim)
    motor_gur = spa.State(dimensions, neurons_per_dimension=n_per_dim)
    
    
    
    
    
    with spa.ActionSelection(normal) as normal_as:
        for elt in sequence:              
            spa.ifmax(spa.dot(vision, spa.sym(elt)),
                    vision >> motor_normal)
    
    with spa.ActionSelection(PD) as PD_as:
        for elt in sequence:              
            spa.ifmax(spa.dot(vision, spa.sym(elt)),
                    vision >> motor_PD)
            
            
            
         
    spa.modules.basalganglia.BasalGanglia.sBCBG = False
    with spa.ActionSelection() as gur_as:
        for elt in sequence:              
            spa.ifmax(spa.dot(vision, spa.sym(elt)),
                    vision >> motor_gur)
    
    normal = {}
    PD = {}
    gur = {}
    
    normal['in'] = [nengo.Probe(pop) for pop in normal_as.bg.input]
    normal['CSN_in'] = [nengo.Probe(pop.neurons, 'input') for pop in normal_as.bg.pops['CSN']]
    normal['CSN'] = [nengo.Probe(pop.neurons) for pop in normal_as.bg.pops['CSN']]
    normal['MSN'] = [nengo.Probe(pop.neurons) for pop in normal_as.bg.pops['MSN']]
    normal['STN'] = [nengo.Probe(pop.neurons) for pop in normal_as.bg.pops['STN']]
    normal['GPe'] = [nengo.Probe(pop.neurons) for pop in normal_as.bg.pops['GPe']]
    normal['GPi'] = [nengo.Probe(pop.neurons) for pop in normal_as.bg.pops['GPi']]
    normal['out'] = [nengo.Probe(pop) for pop in normal_as.bg.output]
    normal['p_motor'] = nengo.Probe(motor_normal.output, synapse=.005)
    
    PD['in'] = [nengo.Probe(pop) for pop in PD_as.bg.input]
    PD['CSN_in'] = [nengo.Probe(pop.neurons, 'input') for pop in PD_as.bg.pops['CSN']]
    PD['CSN'] = [nengo.Probe(pop.neurons) for pop in PD_as.bg.pops['CSN']]
    PD['MSN'] = [nengo.Probe(pop.neurons) for pop in PD_as.bg.pops['MSN']]
    PD['STN'] = [nengo.Probe(pop.neurons) for pop in PD_as.bg.pops['STN']]
    PD['GPe'] = [nengo.Probe(pop.neurons) for pop in PD_as.bg.pops['GPe']]
    PD['GPi'] = [nengo.Probe(pop.neurons) for pop in PD_as.bg.pops['GPi']]
    PD['out'] = [nengo.Probe(pop) for pop in PD_as.bg.output]
    PD['p_motor'] = nengo.Probe(motor_PD.output, synapse=.005)
    
    gur['in'] = [nengo.Probe(pop) for pop in gur_as.bg.input]
    gur['GPi'] = [nengo.Probe(pop.neurons) for pop in gur_as.bg.gpi.ea_ensembles]
    gur['out'] = [nengo.Probe(pop) for pop in gur_as.bg.output]
    gur['p_motor'] = nengo.Probe(motor_gur.output, synapse=.005)
    
    p_vision = nengo.Probe(vision.output, synapse=.01)
    
    
dt = .001
with nengo.Simulator(net, dt=dt) as sim:
    sim.run(.6)
```

## Save results

```python
path='log/'
os.makedirs(path)
np.save(path+'trange', sim.trange())
np.save(path+'vision', spa.similarity(sim.data[p_vision], vocab))
np.save(path+'normal', spa.similarity(sim.data[normal['p_motor']], vocab))
np.save(path+'PD', spa.similarity(sim.data[PD['p_motor']], vocab))
np.save(path+'gur', spa.similarity(sim.data[gur['p_motor']], vocab))
```
